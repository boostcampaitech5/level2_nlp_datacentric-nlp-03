```python
import pandas as pd
import numpy as np
```

```python
crawl_df = pd.read_csv("/opt/level2_nlp_datacentric-nlp-03/crawl_probs.csv")
crawl_df = crawl_df.drop_duplicates(subset=["text"])
crawl_df.head()
```

```python
len(crawl_df)
```

```python
crawl_df["probs"] = crawl_df["probs"].apply(lambda x: np.array(eval(x)))
```

```python
idx_to_text = ["IT과학", "경제", "사회", "생활문화", "세계", "스포츠", "정치"]
text_to_idx = {v: k for k, v in enumerate(idx_to_text)}
```

```python
crawl_df["pred"] = crawl_df["probs"].apply(lambda x: idx_to_text[np.argmax(x)])
crawl_df["max_prob"] = crawl_df["probs"].apply(lambda x: max(x))
```

```python
crawl_df[crawl_df["max_prob"] < 0.4]
```

```python
crawl_train_df = crawl_df[crawl_df["max_prob"] > 0.8]
crawl_train_df["target"] = crawl_train_df["pred"].apply(lambda x: text_to_idx[x])
crawl_train_df = crawl_train_df[["ID", "text", "target"]]
crawl_train_df.sample(10)
```

```python
len(crawl_train_df)
```

```python
crawl_train_df.to_csv("../../data/crawl_train_v2.csv", index=False)
```

```python
crawl_rest_df = crawl_df[crawl_df["max_prob"] <= 0.8]
crawl_rest_df.to_csv("../../data/crawl_rest_v2.csv", index=False)
```
